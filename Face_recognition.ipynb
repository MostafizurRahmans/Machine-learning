```python
from matplotlib import pyplot as plot
```

```python
tramptrudo = cv2.imread("Bangladesh-cricket-team_16a72939a5f_large.jpg")
```

```python
plot.imshow(tramptrudo)
```

```python
from pylab import rcParams
```

```python
def plt_show(image, title="",gray =False,size = (12,10)):
    temp =image
    
    #fix_colour
    if gray == False:
        imagee = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
     
        
    #change image size
    plot.rcParams["figure.figsize"] = size[0],size[1]
    #remove axes ticks
    
    plot.axis("off")
    
    plot.title(title)
    plot.imshow(imagee,cmap='gray')
    plot.show()
```

```python
plt_show(tramptrudo ,"Bangladesh National Team")
```

```python
import urllib.request
harcascade_url = 'https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml'
haaar_name="haarcascade_frontalface_default.xml"

urllib.request.urlretrieve(harcascade_url,haaar_name )
```

```python
detector =cv2.CascadeClassifier('haarcascade_frontalface_defaulta.xml')
```

```python

face_list = detector.detectMultiScale(tramptrudo, scaleFactor=1.1,
minNeighbors=5)

print(face_list)

```

```python
for face in face_list:
    (x ,y ,w , h) = face
    cv2.rectangle(tramptrudo,(x ,y),(x + w,y + h),(0 ,255 ,0),2)
    
```

```python
plt_show(tramptrudo)
```

```python
for face in face_list:
    (x,y,w,h) = face
    
    face =tramptrudo [y:y + h, x:x + w]
    face_resize = cv2.resize(face,(100,100))
```

```python
plt_show( face_resize, size =(3,2))
```
